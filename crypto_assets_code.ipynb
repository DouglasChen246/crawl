```python
import time
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome()

url = "https://www.stakingrewards.com/cryptoassets/?page=" 

for i in range(1,2):
    current_url = url + str(i)
    driver.get(current_url)
    time.sleep(5) 
    res = driver.find_elements("xpath","//div[@class='rt-table']/div[@class='rt-tbody']/div[@class='rt-tr-group']")
    assets = []
    for j in range(0,20):
        assets.append(res[j].text.split('\n'))
    crypto_assets = pd.DataFrame(assets,columns=['#','Asset','Price','24h','Reward','Adj. Reward','Staking Marketcap','Market Cap','Staking Ratio','Add'])
    time.sleep(2)

driver.quit()
```
